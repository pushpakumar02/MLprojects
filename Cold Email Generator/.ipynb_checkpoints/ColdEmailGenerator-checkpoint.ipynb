```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://zoox.com/careers/3021c042-c2e9-495f-8bbe-017792e519cf")
job = loader.load().pop().page_content
print(job)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPE TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scrape text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containg the 
    following keys: 'role', 'experence', 'skills' and 'description'.
    Only return the valid JSON.
    ### VALID JSON (No PREAMBLE):
    """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res 
```

```python
from langchain.prompts import PromptTemplate  # Import PromptTemplate from LangChain

# Define the prompt template
prompt_email = PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}

    ### INSTRUCTION:
    You are Pushpa Kumar Balan who is actively looking for internship opportunities in Software Development and Machine Learning Domain.
    Your job is to write a cold email to the hiring team regarding the job mentioned above describing the capabilities in fulfilling their needs.
    Also add the relevant skills and key words from the following: {link_list}
    Remember you are Pushpa Kumar Balan who is actively looking for internship opportunities in Software Development and Machine Learning Domain.
    Do not provide a PREAMBLE.
    ### EMAIL (NO PREAMBLE):
    """
)

# Chain the email template and LLM (Example placeholder)
# Ensure `llm` is defined, such as using OpenAI's LLM or another model
chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job), "link_list": data})
print(res.content)
```
